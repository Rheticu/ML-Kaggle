```python
X_train_full = pd.read_csv("/kaggle/input/titanic/train.csv")
X_test_full = pd.read_csv("/kaggle/input/titanic/test.csv")
X_train_full.isnull().sum()
X_test_full.isnull().sum()

#X_train_full.head()
```

```python
y_train = X_train_full.Survived
X_train = X_train_full.drop(['Survived','Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
#X_train.isnull().sum()
#X_train.dtypes
X_test = X_test_full.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
#### TUVE QUE QUITAR 'EMBARKING', PERO ESTARÍA BIEN INTENTARLO DESPUÉS CON ÉL.
```

```python
#from sklearn.impute import SimpleImputer
#my_imputer = SimpleImputer()
#imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
#imputed_X_test = pd.DataFrame(my_imputer.fit_transform(X_test))

X_train.Age.fillna(X_train.Age.mean(), inplace=True)
X_test.Age.fillna(X_test.Age.mean(), inplace=True)
X_test.Fare.fillna(X_test.Fare.mean(), inplace=True)

#X_test.isnull().sum()
# Imputation removed column names; put them back
#imputed_X_train.columns = X_train.columns
#imputed_X_test.columns = X_test.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
len(predictions)
```

```python
output = pd.DataFrame({'PassengerId': X_test_full.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
